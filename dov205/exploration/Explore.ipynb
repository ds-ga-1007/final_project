```python
# NOTE: Please run jupyter notebook in the top-level working directory
#       such that pwd = '.../dov205/

import sys
import os

os.chdir('..')

import pandas as pd
import matplotlib.pyplot as plt
import commentary

%matplotlib inline
```

```python
# Read in the data.
data = commentary.load_reddit_data()
```

```python
# Make sure everything looks alright.
data.head()
```

```python
# We have tons of potential directions!
# Let's start by understanding popularity across all subreddits.
popularity = data.groupby(by=data['subreddit']).count()['id'].copy()
popularity.sort_values(inplace=True, ascending=False)

# Barplot of subreddit popularity.
popularity_ax_bar = popularity.plot(title='Subreddit popularity', kind='bar', 
                                    grid=True, figsize=(11, 6))
popularity_ax_bar.set(xlabel = 'Subreddit', ylabel = 'Number of Users')
plt.tight_layout()
plt.savefig("results/subreddit_popularity_bar.png")
plt.show(popularity_ax_bar)
plt.close()
```

```python
# Distribution of subreddit popularity!
popularity_ax_hist = popularity.plot(title='Subreddit popularity', kind='hist', 
                                     grid=True, figsize=(11, 6))
plt.tight_layout()
plt.savefig("results/subreddit_popularity_hist.png")
plt.show(popularity_ax_bar)
plt.close()
```

<ol>

        Despite a fair share of subreddits hovering at approximately 35,000 unique users, we have a relatively long right tail. The subreddits `leagueoflegends` and `funny` are particulary popular, with approximately 150,000 and 134,000 unique users, respectively.
